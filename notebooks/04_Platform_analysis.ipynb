```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(f"F:\Project\Technology Trend\master_clean.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.head()
```

```python
df = df[['PlatformHaveWorkedWith', 'PlatformWantToWorkWith']]
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna()
```

### Count the number of platforms each respondent chose.

```python
df.loc[:, 'count_have_worked'] = df['PlatformHaveWorkedWith'].str.split(';').str.len()
df.loc[:, 'count_want_work'] = df['PlatformWantToWorkWith'].str.split(';').str.len()
df
```

```python
df['count_have_worked'].value_counts().sort_index()
```

```python
df['count_want_work'].value_counts().sort_index()
```

### Only respondents who selected a maximum of 7 platforms have worked with and a maximum of 8 platforms want to work with are considered reliable

```python
df_clean = df[(df['count_want_work']<=8) & (df['count_have_worked']<=7)]
df_clean
```

```python
df_clean.loc[:, 'PlatformHaveWorkedWith'] = df_clean['PlatformHaveWorkedWith'].str.split(';')
df_worked = df_clean['PlatformHaveWorkedWith'].explode().value_counts().reset_index()
df_worked
```

```python
df_worked_t10 = df_worked.head(10)
df_worked_t10
```

```python
df_clean.loc[:, 'PlatformWantToWorkWith'] = df_clean['PlatformWantToWorkWith'].str.split(';')
df_want = df_clean['PlatformWantToWorkWith'].explode().value_counts().reset_index()
df_want
```

```python
df_want_t10 = df_want.head(10)
df_want_t10
```

```python
df_worked_t10_dsc = df_worked_t10.sort_values(by='count', ascending=True)
df_want_t10_dsc = df_want_t10.sort_values(by = 'count', ascending=True)

fig = plt.figure(figsize = (16,7))
ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

bars0 = ax0.barh(df_worked_t10_dsc['PlatformHaveWorkedWith'], 
                 df_worked_t10_dsc['count'], color='r')
ax0.set_title('Top Platform Developers have Worked With')
ax0.set_xlabel('Counts')
ax0.set_ylabel('')
ax0.bar_label(bars0, fmt='%d',padding=-30)

bars1 = ax1.barh(df_want_t10_dsc['PlatformWantToWorkWith'], 
                 df_want_t10_dsc['count'])
ax1.set_title('Top Platform Developers Want to Work With')
ax1.set_xlabel('Counts')
ax1.set_ylabel('')
ax1.bar_label(bars1, fmt='%d', padding= -28)

plt.show()
```

```python
df_worked_t10.to_csv('plworked.csv')
df_want_t10.to_csv('plwant.csv')
```
