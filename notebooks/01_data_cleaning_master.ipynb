### Validate data based on age and working experience

```python
map_exp_age = {'18-24 years old': 6,
               '25-34 years old': 16, 
               '35-44 years old': 26,
               '45-54 years old': 36,
               '55-64 years old': 46,
               '65 years or older': 56,
               'Prefer not to say': 30}
```

```python
df1 = df1.copy()
df1.loc[:,'MaxExp'] = df1['Age'].map(map_exp_age)
```

```python
df1_clean = df1[df1['WorkExp'] <= df1['MaxExp']].drop(columns='MaxExp')
```
