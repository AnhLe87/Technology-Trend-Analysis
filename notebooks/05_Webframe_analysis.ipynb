```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(f"F:\Project\Technology Trend\master_clean.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.head()
```

```python
df.shape
```

```python
df = df[['WebframeHaveWorkedWith', 'WebframeWantToWorkWith']]
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

### Count the number of webframes each respondent chose.

```python
df.loc[:, 'count_have_worked'] = df['WebframeHaveWorkedWith'].str.split(';').str.len()
df.loc[:, 'count_want_work'] = df['WebframeWantToWorkWith'].str.split(';').str.len()
df
```

```python
df['count_have_worked'].value_counts().sort_index()
```

```python
df['count_want_work'].value_counts().sort_index()
```

### Only respondents who selected a maximum of 4 webframes have worked with and a maximum of 5 webframes want to work with are considered reliable

```python
df_clean = df[(df['count_want_work']<=5) & (df['count_have_worked']<=4)]
df_clean
```

```python
df_clean.loc[:, 'WebframeHaveWorkedWith'] = df_clean['WebframeHaveWorkedWith'].str.split(';')
df_worked = df_clean['WebframeHaveWorkedWith'].explode().value_counts().reset_index()
df_worked
```

```python
df_worked_t10 = df_worked.head(10)
df_worked_t10
```

```python
df_clean.loc[:, 'WebframeWantToWorkWith'] = df_clean['WebframeWantToWorkWith'].str.split(';')
df_want = df_clean['WebframeWantToWorkWith'].explode().value_counts().reset_index()
df_want
```

```python
df_want_t10 = df_want.head(10)
df_want_t10
```

```python
df_worked_t10_dsc = df_worked_t10.sort_values(by='count', ascending=True)
df_want_t10_dsc = df_want_t10.sort_values(by = 'count', ascending=True)

fig = plt.figure(figsize = (16,7))
ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

bars0 = ax0.barh(df_worked_t10_dsc['WebframeHaveWorkedWith'], 
                 df_worked_t10_dsc['count'], color='r')
ax0.set_title('Top Webframe Developers have Worked With')
ax0.set_xlabel('Counts')
ax0.set_ylabel('')
ax0.bar_label(bars0, fmt='%d',padding=-30)

bars1 = ax1.barh(df_want_t10_dsc['WebframeWantToWorkWith'], 
                 df_want_t10_dsc['count'])
ax1.set_title('Top Webframe Developers Want to Work With')
ax1.set_xlabel('Counts')
ax1.set_ylabel('')
ax1.bar_label(bars1, fmt='%d', padding= -28)

plt.show()
```

```python
df_worked_t10.to_csv('wfworked.csv')
df_want_t10.to_csv('wfwant.csv')
```
